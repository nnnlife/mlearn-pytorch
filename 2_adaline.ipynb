### Adaptive Linear Neuron (ADALINE)

the percentron was

$
\sigma (z) = 
\begin{cases}
1 \quad if\, z \ge \theta \\
0 \quad otherwise
\end{cases}
$

adaline

$ \sigma (z) = z $

**object function** = lost function or cost function

$ z= w_1x_1 + \dots + w_mx_m + b = w^Tx+b $

Lost function

$ L(w, b)=\frac{1}{2n} \sum_{i=1}^{n} (y^{(i)} - \sigma (z^{(i)}))^2 $

$
\Delta w_j = - \eta \frac{\partial L}{\partial w_j}, \, \Delta b = - \eta \frac{\partial L}{\partial b}
$

$ 
\frac{\partial L}{\partial w_j} = -
\frac{2}{n} \sum_i (y^{(i)} - \sigma (z^{(i)}))x_j^{(i)}
$

$
\frac{\partial L}{\partial b} = -
\frac{2}{n} \sum_i (y^{(i)} - \sigma (z^{(i)}))
$
